Imports:

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

# Opening The Image and showing Informations

```python

img = Image.open("image.png")
plt.imshow(img)
plt.show()
print(img.size)
print(img.format)
print(img.mode)
print(img.has_transparency_data)
width,height = img.size
pixels = img.load()
```

# Spliting Cannels

```python
red,green,blue = img.split()
##built in function

#itterating through the image pixel by pixel and taking only the red color
RedImage = np.zeros((height,width))
for j in range(height):
     for i in range(width):

         RedImage[j,i] = (((pixels[i,j][0]) ))

plt.imshow(red,cmap='Reds')
plt.show()
plt.imshow(green,cmap="Greens")
plt.show()
plt.imshow(blue,cmap="Blues")
plt.show()
plt.imshow(RedImage)
plt.show()
         
```

# Converting in Grey

## First Method 

### Luminosity Method

Grayscale=0.299xR+0.587xG+0.114xB

```python

GrayImage1 = np.zeros((height,width))
for j in range(height):
     for i in range(width):

         GrayImage1[j,i] = ((0.299*(pixels[i,j][0]) + 0.587*(pixels[i,j][1]) + 0.114*(pixels[i,j][2])))
         

plt.imshow(GrayImage1,cmap='gray')
plt.colorbar()
plt.show()
```

the convert function gives the same result as the formula

```python
img = img.convert("L")
plt.imshow(img,cmap='gray')
plt.colorbar()
plt.show()
```

## Second Method


### Naive method
Grayscale = (R+G+B)/3

```python
GrayImage2 = np.zeros((height,width))
for j in range(height):
     for i in range(width):

         GrayImage2[j,i] = (((pixels[i,j][0]) + (pixels[i,j][1]) + (pixels[i,j][2]))/3)
         

plt.imshow(GrayImage2,cmap='gray')
plt.colorbar()
plt.show()
```

## Third Method 

### Decomposition Method

Taking the min or max of the red green or blue

```python
GrayImage3 = np.zeros((height,width))
for j in range(height):
     for i in range(width):

         GrayImage3[j,i] = (min((pixels[i,j][0]),(pixels[i,j][1]),(pixels[i,j][2])))
         

plt.imshow(GrayImage3,cmap='gray')
plt.colorbar()
plt.show()
```

taking The Max

```python
GrayImage4 = np.zeros((height,width))
for j in range(height):
     for i in range(width):

         GrayImage4[j,i] = (max((pixels[i,j][0]),(pixels[i,j][1]),(pixels[i,j][2])))
         

plt.imshow(GrayImage4,cmap='gray')
plt.colorbar()
plt.show()
```
